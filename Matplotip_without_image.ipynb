# Bar graph 

```python
x = [ 1, 2, 3, 4 ]
y = [5, 6, 7, 8]
plt.bar(x,y)
plt.show()
```

```python
x = [ 1, 2, 3, 4 ]
y = [5, 6, 7, 8]
c= ['r','g','b','y']
plt.bar(x,y, color=c)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(["A","B", "C","D"])
y = np.array([3,8,1,10])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(["A","B", "C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y , color ='red') # or color code is will work 
plt.show()
```

# Working with width 

```python
x = np.array(["A","B", "C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y , color ='red', width= 0.2) # or color code is will work 
plt.show()
```

# Working with height

```python
x = np.array(["A","B", "C","D"])
y = np.array([3,8,1,10])
plt.title("Bar") # For Tittle
plt.barh(x,y , color ='red', height = 0.1) # or color code is will work 
plt.show()
```

# Scatter Plot

```python
x = [1, 2,3, 4, 5, 6]
y = [4, 2, 3, 5,1, 8]

plt.title("Scatter plot") # For Tittle 
plt.scatter(x,y)
plt.show()
```

```python
x = [1, 2,3, 4, 5, 6]
y = [4, 2, 3, 5,1, 8]

plt.title("Scatter plot") # For Tittle 
plt.xlabel("Mounth") # X axis
plt.ylabel("Numbers") # Y axis
c= ['red', 'magenta', 'green', 'blue', 'brown','orange']
plt.scatter(x,y, color=c)
plt.show()
```

```python
x = [1, 2,3, 4, 5, 6]
y = [4, 2, 3, 5,1, 8]

plt.title("Scatter plot") # For Tittle 
plt.xlabel("Mounth") # X axis
plt.ylabel("Numbers") # Y axis
c= ['red', 'magenta', 'green', 'blue', 'brown','orange']
plt.scatter(x,y, color=c , s= 150)
plt.show()
```

# Pie Chart 

```python
import matplotlib.pyplot as plt
```

```python
x = [10 , 20 , 30 , 40]
y = ['english', 'science', 'maths' , 'hindi']

plt.pie(x, labels = y)
plt.show()
```

```python
x = [10 , 20 , 30 , 40]
y = ['english', 'science', 'maths' , 'hindi']

c=['yellow','magenta','brown','aqua']
plt.pie(x, labels = y , colors = c)
plt.show()
```

```python
x = [10 , 20 , 30 , 40]
y = ['english', 'science', 'maths' , 'hindi']

plt.pie(x, labels = y)
plt.legend()
plt.show()
```
