```python
# importing Libraries 

import numpy as np
import matplotlib.pyplot as plt 
from PIL import Image

fname = r'logo.png'

# opening Image using PIL 

image = Image.open(fname).convert('L')

# Mapping Image to gray Scale 
plt.imshow(image, cmap ='gray')
plt.show()
```
