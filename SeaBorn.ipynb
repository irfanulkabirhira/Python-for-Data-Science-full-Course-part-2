# Bar Plot in Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("Penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x = df1.island , y=df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island" , y="bill_length_mm", data=df1)
plt.show()
```

```python
df1
```

```python
sns.barplot(x='island' , y ="bill_length_mm", data=df1 , hue = "sex")
plt.show()
```

# Oder 

```python
order1= ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y ="bill_length_mm", data =df1 , hue ="sex" , order =order1)
plt.show()
```

# Hue_Order

```python
order1= ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y ="bill_length_mm", data =df1 , hue ="sex" ,
            order =order1 , hue_order =['FEMALE', 'MALE'])
plt.show()
```

```python
order1 = ["Dream", "Torgerson", "Biscoe"]
sns.barplot(x = 'island', y ='bill_length_mm', data = df1, color ="pink")
plt.show()
```

```python
sns.barplot(x = "island", y ="bill_length_mm", data=df1 , hue="sex", palette="Accent")
plt.show()
```

# Saturation 

```python
sns.barplot(x = "island", y ="bill_length_mm", data=df1 , hue="sex",
            palette="Accent" , saturation =0.2)
plt.show()
```

# Histogram in Seaborn Library 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

# Bins Parameter

```python
sns.displot(df1["flipper_length_mm"], bins =[170 , 180 , 200 , 210 , 220 , 230 , 240])
plt.show()
    
```

# Color

```python
sns.displot(df1["flipper_length_mm"], bins =[170 , 180 , 200 , 210 , 220 , 230 , 240] , color ="g")
plt.show()
    
```

# KDE 

```python
sns.displot(df1["flipper_length_mm"], bins =[170 , 180 ,190, 200 , 210 , 220 , 230 , 240] , color ="g", kde =True)
plt.show()
    
```
